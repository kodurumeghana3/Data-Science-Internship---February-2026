1. Unique Words in a Sentence

```python
# Taking input
sentence = input("Enter sentence: ")

# Splitting sentence into words
words = sentence.split()

# Converting list to set to remove duplicates
unique_words = set(words)

# Printing results
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

2. Highest Salary from Employee Data

```python
# Employee dictionary
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Finding employee with highest salary
highest_paid = max(employees, key=employees.get)

# Printing result
print("Highest Salary:", highest_paid, "-", employees[highest_paid])
```

3. Find Maximum and Minimum Values

```python
# Given list
numbers = [45, 22, 89, 10, 66]

# Initialize max and min with first element
max_value = numbers[0]
min_value = numbers[0]

# Loop through the list
for num in numbers:
    if num > max_value:
        max_value = num
    if num < min_value:
        min_value = num

# Print results
print("List:", numbers)
print("Max:", max_value)
print("Min:", min_value)
```

4. Count Products Above a Price Threshold

```python
# Given list
prices = [450, 1200, 899, 1500, 300]

# Initialize counter
count = 0

# Loop through prices
for price in prices:
    if price > 1000:
        count += 1

# Print result
print("Products above 1000:", count)
```

5. Calculate Attendance Percentage

```python
# Given list
attendance = ["P", "P", "A", "P", "P"]

# Count present days
present_count = 0

for status in attendance:
    if status == "P":
        present_count += 1

# Calculate percentage
percentage = (present_count / len(attendance)) * 100

# Print result
print("Attendance Percentage:", percentage)
```

6. Remove Duplicate Phone Numbers

```python
# Given list
phone_numbers = [9876543210, 9123456789, 9876543210]

# Remove duplicates using set
unique_numbers = set(phone_numbers)

# Print result
print("Unique phone numbers:", unique_numbers)
```

7. Count Character Frequency 

```python
# Given string
text = "python"

# Empty dictionary
char_count = {}

# Loop through each character
for ch in text:
    if ch in char_count:
        char_count[ch] += 1
    else:
        char_count[ch] = 1

# Print result
print(char_count)
```

9. Check if a key exists in a Dictionary

```python
# Given dictionary
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Key to check
key_to_check = "Ravi"

# Checking if key exists
if key_to_check in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```

10. Calculate Average Marks

```python
# Given list
marks = [70, 75, 80, 65]

# Calculate average
average = sum(marks) / len(marks)

# Print result
print("Average Marks:", average)
```
