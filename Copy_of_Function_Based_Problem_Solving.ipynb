1. Social Media-Post Engagement Analyzer


```python
def analyze_post_engagement(likes_list):
    total_likes = 0

    # Loop to calculate total likes
    for like in likes_list:
        total_likes += like

    # Conditional check
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    # Output
    print("Total Likes:", total_likes)
    print("Post Status:", status)


# Example usage
likes = [200, 150, 300, 250, 150]  # Sample likes data
analyze_post_engagement(likes)
```

3.  Agriculture – Rainfall Adequacy Checker



```python
def rainfall_checker(rainfall_list, required_level):

    total_rainfall = 0

    # Loop to calculate total rainfall
    for rain in rainfall_list:
        total_rainfall += rain

    # Calculate average rainfall
    average_rainfall = total_rainfall / len(rainfall_list)

    # Conditional check
    if average_rainfall >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    # Output (rounded for clean display)
    print("Average Rainfall:", round(average_rainfall))
    print("Rainfall Status:", status)


# Example usage
seasonal_rainfall = [70, 75, 68, 80, 67]
required = 70
rainfall_checker(seasonal_rainfall, required)
```

4. Social Media – Duplicate Account Detection



```python
def detect_duplicate_accounts(usernames):

    unique_usernames = set(usernames)  # Set removes duplicates automatically

    # Check for duplicates
    if len(usernames) != len(unique_usernames):
        result = "Yes"
    else:
        result = "No"

    print("Duplicate Accounts Found:", result)


# Example usage
users = ["meghana01", "john23", "meghana01", "alex99"]
detect_duplicate_accounts(users)
```

6.  Agriculture – Premium Crop Price Filter


```python
def filter_premium_crops(price_list):

    premium_crops = []

    # Loop to filter prices
    for price in price_list:
        if price > 2000:
            premium_crops.append(price)

    print("Premium Crops:", premium_crops)


# Example usage
crop_prices = [1500, 2500, 1800, 3200, 1900]
filter_premium_crops(crop_prices)
```

7. System Monitoring – Application Health Checker


```python
def application_health_checker(error_count):

    # Conditional checks
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    # Output
    print("Error Count:", error_count)
    print("System Status:", status)


# Example usage
application_health_checker(7)
```

8.  Banking – Daily Transaction Limit Checker

```python
def check_transaction_limit(amount):

    daily_limit = 50000  # ₹50,000 limit

    # Conditional check
    if amount <= daily_limit:
        status = "Approved"
    else:
        status = "Rejected"

    # Output
    print("Transaction Amount:", amount)
    print("Transaction Status:", status)


# Example usage
check_transaction_limit(60000)
```

9. E-Learning – Student Attendance Eligibility System


```python
def check_attendance_eligibility(attendance_list):

    total_days = len(attendance_list)
    present_days = 0

    # Loop to count present days
    for day in attendance_list:
        if day == "P":   # P = Present
            present_days += 1

    # Calculate attendance percentage
    attendance_percentage = (present_days / total_days) * 100

    # Check eligibility
    if attendance_percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    # Output
    print("Attendance Percentage:", attendance_percentage)
    print("Exam Eligibility:", status)


# Example usage
attendance = ["P", "P", "A", "P", "P"]  # 4 out of 5 days present
check_attendance_eligibility(attendance)
```

10. Smart Electricity Bill Analyzer


```python
def electricity_bill_analyzer(units):

    bill = 0

    # Billing calculation using slabs
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Usage classification
    if bill < 500:
        status = "Low Usage"
    elif 500 <= bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    return bill, status


# Example usage
units_consumed = 250
total_bill, usage_status = electricity_bill_analyzer(units_consumed)

print("Units Consumed:", units_consumed)
print("Total Bill: ₹", total_bill)
print("Usage Level:", usage_status)
```
