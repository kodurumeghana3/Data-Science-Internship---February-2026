```python
from langchain_core.prompts import PromptTemplate, ChatPromptTemplate

print("✅ Libraries loaded")
```

```python
def basic_prompt(topic):
    template = PromptTemplate(
        input_variables=["topic"],
        template="Describe {topic} in an easy way for beginners."
    )
    return template.format(topic=topic)

# Test
print(basic_prompt("Artificial Intelligence"))
```

```python
def multi_prompt(topic, level, style):
    template = PromptTemplate(
        input_variables=["topic", "level", "style"],
        template="Explain {topic} to {level} learners using a {style} explanation."
    )
    return template.format(topic=topic, level=level, style=style)

# Test
print(multi_prompt("Python", "beginners", "simple"))
```

```python
def prompt_modes(topic, mode):

    templates = {
        "guide": PromptTemplate(
            input_variables=["topic"],
            template="Provide a step-by-step guide on {topic}."
        ),
        "quiz": PromptTemplate(
            input_variables=["topic"],
            template="Generate 3 quiz questions on {topic}."
        ),
        "story": PromptTemplate(
            input_variables=["topic"],
            template="Explain {topic} through a short story."
        )
    }

    if mode not in templates:
        return "Invalid mode"

    return templates[mode].format(topic=topic)

# Test
print(prompt_modes("Machine Learning", "guide"))
print(prompt_modes("Machine Learning", "quiz"))
print(prompt_modes("Machine Learning", "story"))
```

```python
def create_chat(role, topic):

    role_map = {
        "teacher": "You are a knowledgeable teacher.",
        "hr": "You are an HR interviewer.",
        "coach": "You guide and motivate learners."
    }

    system_msg = role_map.get(role, "You are a helpful assistant.")

    chat = ChatPromptTemplate.from_messages([
        ("system", system_msg),
        ("user", "Explain {topic}")
    ])

    return chat.format_messages(topic=topic)

# Test
print(create_chat("teacher", "Neural Networks"))
```

```python
def check_inputs(level, tone):

    levels = ["beginner", "intermediate", "advanced"]
    tones = ["formal", "casual", "fun"]

    if level not in levels:
        print("Invalid level → default = beginner")
        level = "beginner"

    if tone not in tones:
        print("Invalid tone → default = formal")
        tone = "formal"

    return level, tone

# Test
print(check_inputs("kids", "fun"))
```

```python
def smart_prompt(topic, level, tone, format_type):

    level, tone = check_inputs(level, tone)

    if format_type == "explanation":
        template = PromptTemplate(
            input_variables=["topic", "level", "tone"],
            template="Explain {topic} to a {level} audience in a {tone} tone."
        )

    elif format_type == "questions":
        template = PromptTemplate(
            input_variables=["topic"],
            template="Create 3 important questions on {topic}."
        )
        return template.format(topic=topic)

    elif format_type == "analogy":
        template = PromptTemplate(
            input_variables=["topic", "level", "tone"],
            template="Explain {topic} using a real-life analogy for {level} learners in a {tone} tone."
        )

    else:
        return "Invalid format"

    return template.format(topic=topic, level=level, tone=tone)

# Test
print(smart_prompt("Neural Networks", "beginner", "fun", "analogy"))
```

```python
def run_system(topic, level, tone, format_type):

    print("="*50)
    print("PROMPT GENERATION SYSTEM")
    print("="*50)

    print("\nInput:")
    print("Topic:", topic)
    print("Level:", level)
    print("Tone:", tone)
    print("Format:", format_type)

    # Validation
    level, tone = check_inputs(level, tone)

    print("\nAfter Validation:")
    print("Level:", level)
    print("Tone:", tone)

    # Generate prompt
    output = smart_prompt(topic, level, tone, format_type)

    print("\nGenerated Prompt:")
    print(output)

    print("\nDone")

    return output

# Run
run_system("Deep Learning", "beginner", "fun", "explanation")
```

```python
test_cases = [
    ("AI", "beginner", "formal", "explanation"),
    ("Blockchain", "advanced", "casual", "analogy"),
    ("Python", "beginner", "fun", "questions"),
    ("Data Science", "intermediate", "formal", "explanation")
]

print("Batch Testing:\n")

for t in test_cases:
    result = smart_prompt(*t)
    print("Input:", t)
    print("Output:", result)
    print("-"*40)
```
