```python
from transformers import pipeline
from langchain_community.llms import HuggingFacePipeline

from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
```

```python
from transformers import pipeline
from langchain_community.llms import HuggingFacePipeline

pipe = pipeline(
    "text-generation",
    model="gpt2",
    max_new_tokens=100,   # ✅ IMPORTANT FIX
    temperature=0.7,
    truncation=True       # ✅ prevents overflow
)

llm = HuggingFacePipeline(pipeline=pipe)
```

```python
print(llm.invoke("Explain Artificial Intelligence in simple terms"))
```

```python
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

prompt = PromptTemplate.from_template(
    "Explain {topic} in simple terms"
)

chain = prompt | llm | StrOutputParser()

print(chain.invoke({"topic": "Machine Learning"}))
```

```python
chat_history = []

def chat(user_input):
    global chat_history

    # keep only last 2 conversations (prevents long input)
    chat_history = chat_history[-2:]

    chat_history.append(f"User: {user_input}")

    context = "\n".join(chat_history)

    response = llm.invoke(context + "\nAI:")

    chat_history.append(f"AI: {response}")

    return response

print(chat("Hi, my name is Meghana"))
print(chat("What is my name?"))
```
