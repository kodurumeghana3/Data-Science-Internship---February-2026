```python
Task 1: Conceptual Understanding

1. What is the difference between "Love" and "love" in NLP?

In NLP, "Love" and "love" are considered different words because computers treat uppercase and lowercase letters separately. This is called case sensitivity. To avoid duplication and improve text analysis, NLP preprocessing usually converts all words to lowercase.

2️. What happens if stopwords are not removed?

If stopwords are not removed, the dataset will contain many common words like “is”, “the”, and “and” that do not add much meaning. This increases the size of the data and processing time. It may also reduce the efficiency of NLP models.

3️. Provide two real-world scenarios where removing stopwords can be harmful **

Removing stopwords can sometimes change the actual meaning of a sentence. For example, in sentiment analysis removing the word “not” from “not good” changes the sentiment. It can also affect question-answer systems, where words like “what” or “who” are important.

4️. Explain the difference between stemming and lemmatization

Stemming reduces a word to its root form by removing suffixes using simple rules, and the result may not always be a real word. Lemmatization converts a word to its correct dictionary base form using linguistic analysis. Lemmatization is usually more accurate but slower than stemming.
```

```python
from collections import Counter
import re

def preprocess_text(text):

    text = text.lower()
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()

    tokens = text.split()

    filtered_tokens = []
    for word in tokens:
        if len(word) > 2 or word in ["no", "not"]:
            filtered_tokens.append(word)

    return filtered_tokens


sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

all_tokens = []

for s in sentences:
    tokens = preprocess_text(s)
    all_tokens.extend(tokens)

word_freq = Counter(all_tokens)

print("Top 10 Most Frequent Words:")
print(word_freq.most_common(10))

print("\nTop 5 Least Frequent Words:")
least_common = word_freq.most_common()[:-6:-1]
print(least_common)
```

```python
import re

def preprocess_text(text):

    if not isinstance(text, str) or text.strip() == "":
        return [], ""

    text = text.lower()
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()

    if text == "":
        return [], ""

    tokens = text.split()

    filtered_tokens = []
    for word in tokens:
        if len(word) > 2 or word in ["no", "not"]:
            filtered_tokens.append(word)

    cleaned_sentence = " ".join(filtered_tokens)

    return filtered_tokens, cleaned_sentence
```
