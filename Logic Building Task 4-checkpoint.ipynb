1. Mobile Recharge Validation System

```python
# Function to validate recharge
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]
    
    if amount < 50:
        print("Recharge amount must be at least ₹50.")
        return False
    
    if amount not in valid_plans:
        print("Invalid plan selected.")
        print("Available plans are:", valid_plans)
        return False
    
    return True


# Retry system using while loop
while True:
    amount = int(input("Enter recharge amount: ₹"))
    
    if validate_recharge(amount):
        print("Recharge Successful ✅")
        break
    else:
        print("Please try again.\n")
```

2. Inventory Reorder Alert System

```python
# Function to check stock levels
def check_inventory(products):
    for product, stock in products.items():
        if stock < 15:
            print(product, ": Reorder Alert ⚠")
        else:
            print(product, ": Stock OK ✅")


# Sample dictionary
inventory = {
    "Rice": 10,
    "Wheat": 25,
    "Sugar": 8,
    "Oil": 20
}

# Function call
check_inventory(inventory)
```

3. Student Result Processing System

```python
# Function to process student result
def process_result(marks):
    total = 0
    
    # Calculate total using loop
    for mark in marks:
        total += mark
    
    average = total / len(marks)
    
    # Decide result
    if average >= 50:
        return "Pass"
    else:
        return "Fail"


# Sample marks list
student_marks = [60, 45, 70, 50]

# Function call
result = process_result(student_marks)

print("Average Marks:", sum(student_marks) / len(student_marks))
print("Result:", result)
```

4. Cab Fare Estimator with Retry Option

```python
# Function to calculate fare
def calculate_fare(distance, peak):
    base_fare = 50
    fare = base_fare + (12 * distance)
    
    if peak == "yes":
        fare = fare + (0.25 * fare)   # 25% extra
    
    return fare


# Retry system
while True:
    distance = float(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ").lower()
    
    total_fare = calculate_fare(distance, peak)
    print("Total Fare: ₹", total_fare)
    
    retry = input("Do you want to calculate again? (yes/no): ").lower()
    if retry != "yes":
        print("Thank you for using Cab Fare Estimator ")
        break
```

5. Employee Attendance Eligibility Checker

```python
# Function to check eligibility
def check_attendance(attendance_list):
    present_count = 0
    
    # Count present days using loop
    for day in attendance_list:
        if day == "P":
            present_count += 1
    
    percentage = (present_count / len(attendance_list)) * 100
    
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


# Sample attendance list
attendance = ["P", "P", "A", "P", "P", "A", "P", "P"]

# Function call
result = check_attendance(attendance)

print("Attendance Percentage:", (attendance.count("P") / len(attendance)) * 100, "%")
print("Eligibility Status:", result)
```

6. Password Strength Checker

```python
# Function to check password strength
def check_password(password):
    has_digit = False
    has_special = False
    
    if len(password) < 8:
        return "Weak Password (Minimum 8 characters required)"
    
    # Loop through each character
    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in "@#$":
            has_special = True
    
    # Check conditions
    if has_digit and has_special:
        return "Strong Password ✅"
    else:
        return "Weak Password (Must contain digit and special character)"


# Example
pwd = input("Enter password: ")
result = check_password(pwd)
print(result)
```
