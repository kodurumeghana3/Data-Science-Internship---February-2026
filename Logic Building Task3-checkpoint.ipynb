1. Employee Performance Bonus Eligibility

```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Step 1: Assume first score is highest
highest = 0

# Step 2: Find highest score
for score in employees.values():
    if score > highest:
        highest = score

# Step 3: Print employees who have highest score
print("Top Performers Eligible for Bonus:")

for name in employees:
    if employees[name] == highest:
        print(name, end=" ")

print("(Score:", highest, ")")
```

2. Search Query Keyboard Analysis

```python
import string

query = "Buy mobile phone buy phone online"

# Step 1: Convert to lowercase
query = query.lower()

# Step 2: Remove punctuation
query = query.translate(str.maketrans('', '', string.punctuation))

# Step 3: Split into words
words = query.split()

# Step 4: Count frequency
frequency = {}
for word in words:
    frequency[word] = frequency.get(word, 0) + 1

# Step 5: Display only words searched more than once
result = {word: count for word, count in frequency.items() if count > 1}

print(result)
```

3. Sensor Data Validation

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# Store valid readings (even numbers) as (index, value) pairs
valid_readings = []

for index, value in enumerate(sensor_readings):
    if value % 2 == 0:   # Check if even
        valid_readings.append((index, value))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

4. Email Domain Usage Analysis

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# Step 1: Count domains
domain_count = {}

for email in emails:
    domain = email.split("@")[1]   # Extract domain
    domain_count[domain] = domain_count.get(domain, 0) + 1

# Step 2: Calculate percentage
total_users = len(emails)

for domain, count in domain_count.items():
    percentage = (count / total_users) * 100
    print(f"{domain}: {percentage:.0f}%")
```

5. Sales Spike Detection

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Step 1: Calculate average
average = sum(sales) / len(sales)

# Step 2: Calculate 30% above average
threshold = average * 1.30

# Step 3: Round threshold down to nearest hundred
threshold = (threshold // 100) * 100

# Step 4: Detect spike days
for i, value in enumerate(sales):
    if value >= threshold:
        print(f"Day {i+1}: {value}")
```

6. Duplicate User Id Detection

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Step 1: Count frequency of each user ID
id_count = {}

for user in user_ids:
    id_count[user] = id_count.get(user, 0) + 1

# Step 2: Display only duplicates (count > 1)
for user, count in id_count.items():
    if count > 1:
        print(f"{user} → {count} times")
```
