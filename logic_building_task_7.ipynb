1. Smart Light Controller


```python
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"

    def control(self, action):
        self.status = action.upper()
        print(self.name, "is", self.status)


# Input
name = input("Light Name: ")
action = input("Action: ")

# Object creation
light = SmartLight(name)
light.control(action)
```

2. Employee ID Card System


```python
class Employee:
    def __init__(self, name, emp_id, dept):
        self.name = name
        self.emp_id = emp_id
        self.dept = dept

    def display(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.dept)


# Input
name = input("Employee Name: ")
emp_id = input("Employee ID: ")
dept = input("Department: ")

# Object creation
emp = Employee(name, emp_id, dept)
emp.display()
```

3.  Mobile Contact Record


```python
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def show(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


# Input
c = Contact(input("Contact Name: "), input("Phone Number: "))
c.show()
```

4. Product Price Tag Generator


```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def show(self):
        print("Product:", self.name)
        print("Price: ₹" + self.price)


# Input
p = Product(input("Product Name: "), input("Price: "))
p.show()
```

5. Movie Rating Display System


```python
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def show(self):
        print("Movie:", self.name)
        print("Rating:", self.rating, "/ 5")


# Input
m = Movie(input("Movie Name: "), input("Rating: "))
m.show()
```

6.Delivery Address Manager


```python
class Delivery:
    def __init__(self, name, address):
        self.name = name
        self.address = address

    def show(self):
        print("Delivery Details")
        print("Customer:", self.name)
        print("Address:", self.address)


# Input
d = Delivery(input("Customer Name: "), input("Address: "))
d.show()
```
